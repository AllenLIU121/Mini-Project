```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#dataset1

dataset1 = pd.read_csv("./dataset_1.csv" )
xi = dataset1.iloc[:100,0]
yi = dataset1.iloc[:,1]
zi = dataset1.iloc[:,2]
dt = 0.02                     
time_t = len(xi)  

X,Y,Z = np.matrix([[0], [0]]),np.matrix([[0], [0]]) ,np.matrix([[0], [0]])    # initialize the optimized value
X_mat,Y_mat,Z_mat = np.zeros(time_t),np.zeros(time_t),np.zeros(time_t)        # initialize the covariance 
                            
A = np.matrix([[1, dt], [0, 1]])        
P = np.matrix([[1, 0], [0, 1]])                # initialize the covariance matrix
Q = np.matrix([[dt ** 2 / 2], [dt]]) * np.matrix([[dt ** 2 / 2, dt]])       
H = np.matrix([[1, 0]])                      


# x-translation

kf = 2
for i in range(len(xi)):
    R = Q[0,0] * kf
    X_predict = A * X    # Estimate state variables
    P_Xpredict = A * P * A.T + Q             # Estimated state error covariance
    K = P_Xpredict * H.T / (H * P_Xpredict * H.T + R)     # Update the Kalman gain
    X = X_predict + K * (xi[i] - H * X_predict)        # Update forecast optimization value
    P = (P_Xpredict - K * H) * P_Xpredict                # Update state error covariance
    X_mat[i] = X[0, 0]

plt.plot(X_mat)
plt.xlabel("x translation")
plt.show()


# y-translation
kf = 60
for j in range(len(yi)):
    R = Q[0,0] * kf
    Y_predict = A * Y   
    P_Ypredict = A * P * A.T + Q            
    K = P_Ypredict * H.T / (H * P_Ypredict * H.T + R)    
    Y = Y_predict + K * (yi[j] - H * Y_predict)       
    P = (P_Ypredict - K * H) * P_Ypredict                
    Y_mat[j] = Y[0, 0]

plt.plot(Y_mat, "g")
plt.xlabel("y translation")
plt.show()


# z-translation

kf = 0.5
for k in range(len(yi)):
    R = Q[0,0] * kf
    Z_predict = A * Z   
    P_Zpredict = A * P * A.T + Q           
    K = P_Zpredict * H.T / (H * P_Zpredict * H.T + R)    
    Z = Z_predict + K * (zi[k] - H * Z_predict)    
    P = (P_Zpredict - K * H) * P_Zpredict                

    Z_mat[k] = Z[0, 0]


plt.plot(Z_mat, "r")
plt.xlabel("z translation")
plt.show()
```
